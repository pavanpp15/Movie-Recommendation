```python
import pandas as pd
df_movies = pd.read_csv('C:\\Users\\patel\\Desktop\\Pavan Stevens\\Spring 2020 (Sem-2)\\CPE-695 Applied Machine Learning\\Project\\Data\\movies.csv')
df_movies.head()
```

```python
df_ratings = pd.read_csv('C:\\Users\\patel\\Desktop\\Pavan Stevens\\Spring 2020 (Sem-2)\\CPE-695 Applied Machine Learning\\Project\\Data\\ratings.csv')
df_ratings.head()
```

```python
merged_data = pd.merge(df_movies, df_ratings, on='movieId')
merged_data = merged_data.drop('timestamp', axis=1)
merged_data.head()
```

```python
df_link = pd.read_csv('C:\\Users\\patel\\Desktop\\Pavan Stevens\\Spring 2020 (Sem-2)\\CPE-695 Applied Machine Learning\\Project\\Data\\links.csv')
df_link.head()
```

```python
df_link = df_link[['movieId','imdbId']]
df_link.head()
```

```python
merged_data = pd.merge(merged_data,df_link, on='movieId')
merged_data.head()
```

```python
merged_data = merged_data[['userId','movieId','imdbId','title','rating','genres']]
merged_data.head()
```

```python
imdbId_only = merged_data[['imdbId']]
imdbId_only.head()
```

```python
imdbId_only = imdbId_only.drop_duplicates('imdbId')
imdbId_only = imdbId_only.reset_index(drop=True)
imdbId_only.head()
```

```python
import numpy as np

imdbId_array = imdbId_only.to_numpy().flatten()
imdbId_array
```

```python
from imdb import IMDb
fa = IMDb()
```

```python
def findDirector():
    info_Matrix = fa.get_movie(i)
    for director in info_Matrix['director']:
        return director['name']
    
```

```python
director_array=[]
for i in imdbId_array:
    try:
        x = findDirector()
        director_array.append(x)
    except:
        director_array.append('0')
        
        
```

```python
director_array
```

```python
len(director_array)
```

```python
imdbId_only['Director'] = director_array
imdbId_only.head()
```

```python
merged_all_data = pd.merge(merged_data,imdbId_only, on ='imdbId')
merged_all_data.head()
```

```python
merged_all_data.to_csv('C:\\Users\\patel\\Desktop\\Pavan Stevens\\Spring 2020 (Sem-2)\\CPE-695 Applied Machine Learning\\Project\\Data\\merged_all_data.csv')
```

```python
def findPlot():
    the_matrix = fa.get_movie(i)
    for plot in the_matrix['plot']:
        return plot
```

```python
plot_array = []
```

```python
for i in imdbId_array:
    try:
        x = findPlot()
        plot_array.append(x)
    except:
        plot_array.append('0')
```

```python
imdbId_only['Plot'] = plot_array
```

```python
imdbId_only.head()
```

```python
final_merged_all_data = pd.merge(merged_all_data,imdbId_only[['imdbId','Plot']], on = 'imdbId')
```

```python
final_merged_all_data.head()
```

```python
final_merged_all_data.to_csv('C:\\Users\\patel\\Desktop\\Pavan Stevens\\Spring 2020 (Sem-2)\\CPE-695 Applied Machine Learning\\Project\\Data\\final_merged_all_data.csv')
```

```python
def findCast():
    the_matrix = fa.get_movie(i)
    for cast in the_matrix['cast']:
        return cast['name']
```

```python
cast_array = []
```

```python
for i in imdbId_array:
    try:
        x = findCast()
        cast_array.append(x)
    except:
        cast_array.append('0')
```

```python
imdbId_only['Cast'] = cast_array
```

```python
final_data = pd.merge(final_merged_all_data, imdbId_only, on='imdbId')
```
